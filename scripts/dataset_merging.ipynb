```python
import pandas as pd
import numpy as np
import tensorflow
```

```python
customer_profile = pd.read_csv("dataset/customer_social_profiles - customer_social_profiles.csv")
customer_transactions = pd.read_csv("dataset/customer_transactions - customer_transactions.csv")
customer_profile.head()
```

```python
customer_transactions.head()
```

```python
# clean customer_profile dataset
customer_profile['customer_id_new'] = customer_profile['customer_id_new'].astype(str).str.replace(r'^A', '', regex=True)
customer_profile['customer_id_new'] = customer_profile['customer_id_new'].astype(int)
customer_profile.head()
```

```python
# merge datasets
merged_df = pd.merge(customer_profile, customer_transactions, left_on='customer_id_new', right_on='customer_id_legacy', how='inner')
merged_df.drop(columns=['customer_id_legacy'], inplace=True)
```
